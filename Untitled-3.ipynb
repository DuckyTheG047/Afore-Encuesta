```python
import pandas as pd
import zipfile
from pathlib import Path

base_path = Path("/Users/patoescamilla/Desktop/Files/Python/ENIGH")

with zipfile.ZipFile(base_path / "enigh2024_ns_poblacion_csv.zip", "r") as z:
    poblacion = pd.read_csv(z.open(z.namelist()[0]))

with zipfile.ZipFile(base_path / "enigh2024_ns_concentradohogar_csv.zip", "r") as z:
    hogar = pd.read_csv(z.open(z.namelist()[0]))

with zipfile.ZipFile(base_path / "enigh2024_ns_trabajos_csv.zip", "r") as z:
    trabajo = pd.read_csv(z.open(z.namelist()[0]))

with zipfile.ZipFile(base_path / "enigh2024_ns_viviendas_csv.zip", "r") as z:
    viviendas = pd.read_csv(z.open(z.namelist()[0]))

with zipfile.ZipFile(base_path / "enigh2024_ns_gastoshogar_csv.zip", "r") as z:
    gastohogar = pd.read_csv(z.open(z.namelist()[0]))

with zipfile.ZipFile(base_path / "enigh2024_ns_gastotarjetas_csv.zip", "r") as z:
    gastotarjetas = pd.read_csv(z.open(z.namelist()[0]))

with zipfile.ZipFile(base_path / "enigh2024_ns_ingresos_csv.zip", "r") as z:
    ingresos = pd.read_csv(z.open(z.namelist()[0]))  

```

```python
gastotarjetas
```

```python
df = poblacion.merge(
    trabajo,
    on=["folioviv", "foliohog", "numren"],
    how="left"
)

df = df.merge(
    hogar,
    on=["folioviv", "foliohog"],
    how="left"
)

df = df.merge(
    viviendas[["folioviv", "tam_loc"]],
    on="folioviv",
    how="left"
)

df = df.merge(
    ingresos[["folioviv", "foliohog", "numren"]],
    on=["folioviv", "foliohog", "numren"],
    how="left"
)

df = df.merge(
    gastohogar[["folioviv", "foliohog", "clave"]],
    on=["folioviv", "foliohog"],
    how="left"
)

df = df.merge(
    gastotarjetas[["folioviv", "foliohog", "gastotarjetas"]],
    on=["folioviv", "foliohog"],
    how="left"
)

#df = df[
#    (df["edad"].between(18, 27)) &
#    (df["htrab"] > 0)
#]

df["urbano"] = (df["tam_loc"] <= 2).astype(int)


df["nivelaprob_num"] = pd.to_numeric(df["nivelaprob"], errors="coerce")

df["educ_media_sup"] = (df["nivelaprob_num"] >= 6).astype(int)
```
